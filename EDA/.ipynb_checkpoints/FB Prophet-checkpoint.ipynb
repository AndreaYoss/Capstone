```python
%matplotlib inline
import pandas as pd
from fbprophet import Prophet

import matplotlib.pyplot as plt


import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

import seaborn as sns
%matplotlib inline

from sklearn.model_selection import train_test_split

from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

import pmdarima as pm

# This will allow us to avoid a FutureWarning when plotting.
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
daily = pd.read_csv('../Data/S00-USA-csv.csv')
daily.head()
```

```python
daily['Date'] = pd.to_datetime(daily['Date'])
#daily.set_index('ds', inplace=True)
#daily.sort_index(inplace=True)
#daily['month'] = daily.index.month
#daily['day'] = daily.index.day
#daily['year'] = daily.index.year
```

```python
daily.head()
```

```python
daily.drop(columns = ['Last','Change', '% Change', 'Bid', 'Ask',
       'Open Interest', 'Volume'], inplace = True)
```

```python
daily.columns
```

```python
daily.isnull().sum()
```

```python
daily = daily.rename(columns = {'Date': 'ds',
                               'Settlement Price': 'y'})
```

```python
daily.head()
```

```python
daily.sort_index(ascending = False, inplace = True)
```

## FB Prophet

```python
model_train = Prophet()
```

```python
rows_train = round(0.85*len(daily))
rows_test = len(daily) - rows_train
train = daily.head(rows_train)
test = daily.tail(rows_test)
```

```python
train_ind = train.set_index('ds')
```

```python
test_dates = pd.DataFrame(test['ds'])
test_dates
```

```python
model_train.fit(train)
```

```python
forecasts = model_train.predict(test_dates)[['ds', 'yhat']]
type(forecasts)
```

```python
forecasts.describe()
```

```python
plt.figure(figsize = (10,10))
plt.plot(forecasts['ds'], forecasts['yhat'])
```

```python
forecasts['y']=test['y']
```

```python
forecasts.set_index('ds', inplace = True)
```

```python
mean_absolute_error(forecasts['y'], forecasts['yhat'])
```

```python
forecasts.plot()
```

```python
resid = forecasts['y']-forecasts['yhat']
```

```python
# Plot data.
plt.figure(figsize=(10,6))
plt.plot(resid, 'bo', color = 'blue')
plt.ylabel('Soybean Daily Price (USD)')
plt.xlabel('Date')
plt.title(label = 'Residual Plot', fontsize=16)
plt.show();
```

```python
model_season = Prophet(
    #growth='linear',
    #changepoints=None,
    #n_changepoints=25,
    #changepoint_range=0.8,
    yearly_seasonality='auto',
    weekly_seasonality='auto',
    daily_seasonality='auto',
    holidays=None,
    seasonality_mode='additive',
    seasonality_prior_scale=20.0,
    holidays_prior_scale=10.0,
    changepoint_prior_scale=0.05,
    mcmc_samples=0,
    interval_width=0.8,
    uncertainty_samples=1000,
    stan_backend=None,
)
```

```python
model_season.fit(train)
```

```python
forecasts_season = model_season.predict(test_dates)[['ds', 'yhat']]
type(forecasts_season)
```

```python
forecasts_season.describe()
```

```python
plt.figure(figsize = (10,10))
plt.plot(forecasts_season['ds'], forecasts_season['yhat'])
```

```python
forecasts_season['y']=test['y']
forecasts_season.set_index('ds', inplace = True)
```

```python
mean_absolute_error(forecasts_season['y'], forecasts_season['yhat'])
```

by increasing the impact of seasonality, the model actually did worse!

```python
model_V3 = Prophet(seasonality_mode='multiplicative')
```

```python
model_V3.fit(train)
```

```python
forecasts_V3 = model_V3.predict(test_dates)[['ds', 'yhat']]
forecasts_V3['y']=test['y']
forecasts_V3.set_index('ds', inplace = True)
```

```python
mean_absolute_error(forecasts_V3['y'], forecasts_V3['yhat'])
```

Okay, multilpicative is better than additive for the seasonal component.

```python
forecasts_V3.plot()
```

```python
train['ds']
```

```python
# Plot data.
plt.figure(figsize=(10,6))

# Plot training data.
plt.plot(train_ind.index, pd.DataFrame(train_ind), color = 'blue')

# Plot testing data.
plt.plot(forecasts_V3.index, forecasts_V3['y'], color = 'orange')

# Plot predicted test values.
plt.plot(forecasts_V3.index, forecasts_V3['yhat'], color = 'green')

plt.title(label = 'Daily Average Soybean Price Predictions', fontsize=16)
```
