## Cleaning

### Time Series Dataframe

Reading in daily soybean market price dataframe.

```python
daily = pd.read_csv('../Data/S00-USA-csv.csv')
daily.head()
```

Checking the variable datatypes and if there are any missing values.

```python
daily.info()
```

Some of the variable datatypes are not correct and there appear to be missing values. Since we will be performing time series analyses on the data, I am converting the 'DATE' column to datetime format, setting it as the index, and sorting the dataframe ascending by date. 

```python
daily['Date'] = pd.to_datetime(daily['Date'])
daily.set_index('Date', inplace=True)
daily.sort_index(inplace=True)
daily
```

Evaluating and removing null values.

```python
daily.isnull().sum().sort_values(ascending=False)
```

```python
daily.isnull().sum().sort_values(ascending = False).plot(kind= 'bar')
plt.title('Counts of Null Values in Data')
plt.xlabel('Variables')
plt.ylabel('Null Counts');
```

For the purpose of this analysis, we are only concerned with the Daily Settlement Price, which has 0 null values. We will remove the additional columns from our 'daily' dataframe.

```python
daily = pd.DataFrame(daily['Settlement Price'])
```

```python
daily.to_csv('../Data/settlement_price_df.csv')
```

### Exogenous Features
#### Weather Data
Importing NOAA weather data for the top 3 US soybean producing states.
  
Reading in Iowa data.

```python
IA = pd.read_csv('../Data/NOAA_Iowa.csv')
IA.head()
```

```python
IA.columns
```

```python
IA_prcp = IA[['DATE', 'PRCP', 'TMIN', 'TMAX']]
```

```python
IA_prcp.info()
```

Besides from 'DATE,' the variable datatypes are all correct. There do appear to be missing values which we will want to examine and remove. Since we will be performing time series analyses on the data, I am converting the 'DATE' column to datetime format, setting it as the index, and sorting the dataframe ascending by date. 

```python
IA_prcp['DATE'] = pd.to_datetime(IA_prcp['DATE'])
IA_prcp.set_index('DATE', inplace=True)
IA_prcp.sort_index(inplace=True)
IA_prcp
```

We will ultimately be concatanating this Iowa data with the other state data, so we will rename the columns to make more specific to this data.

```python
IA_prcp.rename(columns = {'PRCP':'IA_PRCP',
                         'TMIN':'TMIN_IA',
                         'TMAX': 'TMAX_IA'}, inplace = True)
```

```python
IA_prcp.columns
```

Reading in Minnesota and Illinois data.

```python
MN_IL = pd.read_csv('../Data/NOAA_MN_IL.csv')
MN_IL.head()
```

```python
MN_IL.columns
```

We will create a new dataframe with features of interest, keeping the 'Station' column in so we can ultimately use that information to differentiate between IL and MN data. Also, converting the 'DATE' to datetime and setting as index.

```python
MN_IL_prcp = MN_IL[['DATE', 'PRCP', 'STATION', 'TMIN', 'TMAX']]
MN_IL_prcp.head()
```

```python
MN_IL_prcp['DATE'] = pd.to_datetime(MN_IL_prcp['DATE'])
MN_IL_prcp.set_index('DATE', inplace=True)
MN_IL_prcp.sort_index(inplace=True)
MN_IL_prcp
```

Creating seperate dataframes for the MN and IL data with renamed columns.

```python
MN_prcp = MN_IL_prcp[MN_IL_prcp['STATION']=='USC00215638'].drop(columns = ['STATION'])
MN_prcp.rename(columns = {'PRCP':'MN_PRCP',
                          'TMIN':'TMIN_MN',
                         'TMAX': 'TMAX_MN'}, inplace = True)
IL_prcp = MN_IL_prcp[MN_IL_prcp['STATION']=='USW00013802'].drop(columns = ['STATION'])
IL_prcp.rename(columns = {'PRCP':'IL_PRCP',
                         'TMIN':'TMIN_IL',
                         'TMAX': 'TMAX_IL'}, inplace = True);
```

Creating a new dataframe by concatenating the data from IL, MN, and IA. I am grouping by 'DATE' and summing so dates will line up no matter the index and any nulls will just become 0.

```python
data = pd.concat([MN_prcp, IL_prcp, IA_prcp]).groupby('DATE').sum()
data
```

Adding the Settlement Price to this dataframe.

```python
data['Settlement Price'] = daily['Settlement Price']
```

```python
data.info()
```

The only nulls in the data are values of the settlement price for days that are not trading days, and thus have no information.  We will drop these.

```python
data.dropna(inplace = True)
```

```python
data.info()
```

```python
data.to_csv('../data/data_concat.csv')
```
